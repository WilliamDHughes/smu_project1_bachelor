```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.width", 120)

```

```python
Bachelor_1 = "../Misc Code/csvfile.csv"
df = pd.read_csv(Bachelor_1)
df
```

```python
df.info()
```

```python
df["Age Bin"]=None
df.loc[(df.Age >= 20) & (df.Age <= 25), "Age Bin"] = "20-25"
df.loc[(df.Age > 25) & (df.Age <= 29), "Age Bin"] = "26-29"
df.loc[(df.Age > 29) & (df.Age <= 35), "Age Bin"] = "30-35"
df.loc[(df.Age > 35) & (df.Age <= 39), "Age Bin"] = "36-39"
df.head()
```

```python
df["Season Bin"]=None
df.loc[(df.Season <= 7),  "Season Bin"] = "01-07"
df.loc[(df.Season >7) & (df.Season <= 14), "Season Bin"] = "08-14"
df.loc[(df.Season > 14),  "Season Bin"] = "15-21"
df.head()
```

```python
df2 = df.groupby(['Season'])[['Age']].agg(['mean', 'count', 'min', 'max'])
df2.info()
```

```python
df3 = df.groupby(['Season'])[['Age']].agg(['mean', 'count'])
df3.head()
```

```python
df3.plot()
```

```python
colors={"mean":"red","count":"pink"}
labels=list(colors.keys())
Season_age_chart = df3.plot(figsize=(15,6),kind="bar", title="Average Age in each Bachelor Season",color=colors.values())
Season_age_chart.set_title("Average Age in each Bachelor Season",fontweight='bold',fontsize=15)
Season_age_chart.set_xlabel("Season",fontweight='bold')
Season_age_chart.set_ylabel("Age",fontweight='bold')

Season_age_chart.set_ylim(20,35)
plt.legend(labels)
plt.show()
plt.tight_layout()
```

```python
colors={"mean":"hotpink","count":"pink"}
labels=list(colors.keys())
sns.set_theme(style="darkgrid")
Season_age_chart = df3.plot(figsize=(15,6),kind="line", color=colors.values(),linewidth=6)
Season_age_chart.set_ylabel("Age",fontdict={'fontsize':'14','fontweight':'bold','color':'red'})
Season_age_chart.set_xlabel("Season",fontdict={'fontsize':'14','fontweight':'bold','color':'red'})
Season_age_chart.set_ylim(20,35)
Season_age_chart.xaxis.set_tick_params(labelsize=15)
Season_age_chart.yaxis.set_tick_params(labelsize=15)
Season_age_chart.set_title("Count and Average Age in each Bachelor Season",fontdict={'fontsize':'14','fontweight':'bold','color':'red'})
plt.legend(loc="upper left",prop={'size': 13,'weight':'bold'})
plt.show()
plt.tight_layout()


```

```python
# step 1, get the data
df8=df.loc[df["Season Bin"]=="15-21"]
data=df8["Age Bin"].value_counts()

# for customization
colors={"30-35":"Red","26-29":"hotpink",'20-25':"magenta"}
labels=list(colors.keys())


# step 2, init the plot
fig = plt.figure(figsize=(10,6))
plt.pie(data.values,labels=data.index, autopct="%1.1f%%", pctdistance=0.85, startangle=140,colors=colors.values(),textprops={'fontsize': 14,'weight':'bold','fontstyle':'italic'})

# step 2.5, draw a circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# step 3 - customization
plt.title("Contestents Age Bins for Season 15-21", fontweight="bold",fontsize='16',color='purple')
plt.axis("equal")

# step 4 - show the pie
plt.show()
```
